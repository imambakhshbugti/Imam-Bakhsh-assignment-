### 1Ô∏è‚É£ Load the Titanic dataset and check missing values

```

import seaborn as sns
import pandas as pd

# Load Titanic dataset from seaborn
data = sns.load_dataset('titanic')

# Show first few rows
print(data.head())

# Count missing values
print("\nMissing values in each column:")
print(data.isnull().sum())
```

**Discussion:**  
- Columns like **age** and **embarked** have few missing values.  
- Column **deck (cabin)** has many missing values, so it can be dropped.  
- Other columns are complete and can be used safely.

### 2Ô∏è‚É£ Imputation by 'Sex' groups instead of 'Pclass'

```

# Fill missing age values by average age in each Sex group
data['age'] = data.groupby('sex')['age'].transform(lambda x: x.fillna(x.mean()))

# Check if any age value is still missing
print("Missing Age values after imputation:", data['age'].isnull().sum())
```

**Explanation:**  
- This fills missing **age** values with the average age of each **sex group**.  
- It is more realistic than using Pclass because men and women have different average ages.

### 3Ô∏è‚É£ Extract and Create a New Feature: 'Title' from 'Name'

```

# Create a new column 'title' from 'name'
data['title'] = data['who']  # using 'who' because seaborn dataset does not have exact name column

# Show few examples
print(data[['sex', 'age', 'class', 'title']].head())
```

**Note:** The seaborn Titanic dataset does not include full passenger names, so here we use the **'who'** column as a simple title (man, woman, child).

### 5Ô∏è‚É£ Final Summary


- Missing values found in *age*, *embarked*, *deck* columns.  
- *Age* and *Embarked* can be filled (imputed).  
- *Deck* can be dropped because too many values are missing.  
- Imputation by *Sex* gives better and realistic results.  
- New feature *Title* helps the model understand gender, age, and social class.
